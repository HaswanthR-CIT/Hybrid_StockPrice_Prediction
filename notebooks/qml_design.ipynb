# QML Component Design

## 1. Import Libraries

```python
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt
```

## 2. Define Quantum Device
Using `default.qubit` with 4 wires.

```python
n_qubits = 4
dev = qml.device('default.qubit', wires=n_qubits)
```

## 3. Design Variational Quantum Circuit (VQC)
Feature extraction using `AngleEmbedding` and `BasicEntanglerLayers`.

```python
@qml.qnode(dev)
def vqc_circuit(inputs, weights):
    # Embed encoded input data
    qml.AngleEmbedding(inputs, wires=range(n_qubits))
    # Trainable VQC layers
    qml.BasicEntanglerLayers(weights, wires=range(n_qubits))
    # Measure expectation values
    return [qml.expval(qml.PauliZ(w)) for w in range(n_qubits)]

# Check circuit structure
dummy_inputs = np.random.random(n_qubits)
dummy_weights = np.random.random((2, n_qubits))
print(qml.draw(vqc_circuit)(dummy_inputs, dummy_weights))
```
